## Keras概述、线性回归

```python
import tensorflow as tf
```

```python
print('TensorFlow Version: {}'.format(tf.__version__))
```

```python
import pandas as pd
```

```python
data = pd.read_csv('dataset/02/Incom1.csv')
```

```python
data
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(data.Education, data.Income)
```

```python
x = data.Education
y = data.Income
```

```python
model = tf.keras.Sequential()
```

Dense(输出维度, 输入维度)

```python
model.add(tf.keras.layers.Dense(1, input_shape=(1,)))
```

Output Shape: (None, 1): None:第一个维度代表输入样本的数量；1代表输入数据维度为1

Param：参数为2

```python
model.summary()
```

编译

```python
model.compile(
    optimizer='adam',
    loss='mse'
)
```

```python
history = model.fit(x, y, epochs=5000)
```

```python
model.predict(x)
```

```python
model.predict(pd.Series([20]))
```
